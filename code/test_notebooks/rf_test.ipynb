```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_x = np.random.normal(0, 1, size = (25, 100))
train_y = np.random.normal(0,1, size = 25)
test_x = np.random.normal(0, 1, size = (25, 100))
```

```python
regr = RandomForestRegressor(max_depth=2, random_state=0)
```

```python
regr.fit(train_x, train_y)
```

```python
regr.predict(test_x)
```

```python
#better for longer segments not individual 'sites' 
#see https://scikit-learn.org/stable/modules/generated/sklearn.inspection.permutation_importance
importances = regr.feature_importances_
importances
```

```python
#plot importances, better for segments
forest_importances = pd.Series(importances)

fig, ax = plt.subplots()
forest_importances.plot.bar(ax=ax)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```
