```python
import easypheno
import os
import pathlib
import numpy as np
import pandas as pd
```

```python
genos = pd.read_csv("data/test_data/genos_reduce.csv")
```

```python
genos
```

```python
genotypes = genos.iloc[:,1:50].to_numpy()
SNP_ids = genos.columns[1:].to_numpy()
sample_ids = genos.iloc[:,0].to_numpy()
```

```python
sim_pheno = easypheno.simulate.synthetic_phenotypes.get_simulation(X = genotypes, sample_ids = sample_ids, snp_ids = SNP_ids,
                                                      number_of_samples = 50, number_causal_snps = 20,
                                                      explained_variance = 80, maf = 30, heritability = 80,
                                                      seed = 42, number_background_snps = 16, distribution = "normal",
                                                      shape = 0.5)
```
